# Generaci√≥ per a N=15

```python
from simglucose.simulation.user_interface import simulate
from simglucose.simulation.scenario import CustomScenario
from datetime import datetime, timedelta
import pandas as pd
from simglucose.controller.basal_bolus_ctrller import BBController
import random

start_time = datetime.now().replace(hour=0, minute=0, second=0, microsecond=0)

adults = ['adult#001', 'adult#002', 'adult#004', 'adult#005', 'adult#008','adult#009', 'adult#010']

meal_times = {'breakfast': 8, 'lunch': 13, 'dinner': 20}  # hora i marge d'error en hores
minute_variation = 60  # Variaci√≥ m√†xima en minuts
carb_ranges = {'breakfast': (40, 60), 'lunch': (70, 90), 'dinner': (60, 80)}  # rangs de carbohidrats per √†pat

def generate_random_meals(start_time):
    meals = []
    for meal, hour in meal_times.items():
        meal_time = start_time + timedelta(hours=hour) + timedelta(minutes=random.randint(-minute_variation, minute_variation))
        
        carb_min, carb_max = carb_ranges[meal]
        meal_carbs = random.randint(carb_min, carb_max)*1.0
        
        meals.append((meal_time, meal_carbs))
    return meals

num_simulations = 15
all_results = []

for i in range(num_simulations):
    random_meals = generate_random_meals(start_time)
    scenario = CustomScenario(start_time=start_time, scenario=random_meals)
    
    cgm_seed = random.randint(0, 1000)
    
    results = simulate(animate=False,
                       parallel=False,
                       sim_time=timedelta(days=1),
                       patient_names=adults,
                       cgm_name='Dexcom',
                       controller=BBController(),
                       insulin_pump_name='Insulet',
                       cgm_seed=cgm_seed,
                       scenario=scenario,
                       start_time=start_time,
                       save_path=f'/Users/danasour/PycharmProjects/GLUSAP/notebooks/results/simulation_{i}')  
    
```

# Processament de les dades

```python
import os

input_path = '/Users/danasour/PycharmProjects/GLUSAP/notebooks/results_processed'
output_path = 'results_grouped'
os.makedirs(output_path, exist_ok=True)


adults = ['adult#001', 'adult#002', 'adult#004', 'adult#005', 'adult#008', 'adult#009', 'adult#010']

target_length = 120  

def process_meal_data(adult):
    grouped_data = {}

    csv_file = os.path.join(input_path, f"{adult}_meals.csv")
    if not os.path.exists(csv_file):
        print(f"No s'ha trobat el fitxer: {csv_file}")
        return

    try:
        df = pd.read_csv(csv_file)
    except Exception as e:
        print(f"Error al llegir el fitxer {csv_file}: {e}")
        return

    if not all(col in df.columns for col in ['Simulation_ID', 'label_meal']):
        print(f"El fitxer {csv_file} no cont√© les columnes necess√†ries.")
        return

    grouped_meals = df.groupby(['Simulation_ID', 'label_meal'])

    for (simulation_id, meal_label), meal_group in grouped_meals:
        meal_subseries = regularize_length(meal_group, target_length)
        
        print(len(meal_subseries))

        if meal_label not in grouped_data:
            grouped_data[meal_label] = []
        grouped_data[meal_label].append(meal_subseries)

    for meal_label, subseries in grouped_data.items():
        if subseries:  
            final_df = pd.concat(subseries, ignore_index=True)
            output_file = os.path.join(output_path, f"{adult}_meals_label_{meal_label}.csv")
            final_df.to_csv(output_file, index=False)
            print(f"S'ha generat el fitxer: {output_file}")
    return grouped_data

def regularize_length(subseries, target_length):
    l = len(subseries)
    if l < target_length:
        add_rows = target_length - l
        subseries = subseries.reset_index(drop=True)
        rows = pd.DataFrame(columns=subseries.columns,
                                index=range(l, target_length),
                                data=[[pd.NA] * len(subseries.columns)] * add_rows)
        subseries = pd.concat([subseries, rows], ignore_index=True)
    elif l > target_length:
        subseries = subseries.iloc[:target_length].reset_index(drop=True)
    else:
        pass
    
    return subseries

for adult in adults:
    grouped_data = process_meal_data(adult)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from glob import glob
import os

adults = ['adult#001']
meal_names = [0, 1, 2] 
data_folder = '/Users/danasour/PycharmProjects/GLUSAP/notebooks/results_grouped'

for adult in adults:
    for meal in meal_names:
        simulation_file = os.path.join(data_folder, f"{adult}_meals_label_{meal}.csv")
        
        concatenated_data = pd.DataFrame()
        current_time_offset = pd.Timedelta(0)  
        
        df = pd.read_csv(simulation_file)
        
        simulations = df['Simulation_ID'].unique()
        
        for sim in simulations:
            sim_data = df[df['Simulation_ID'] == sim]
            
            # Afegeix l'offset de temps a les dades
            sim_data['Time'] = pd.to_datetime(sim_data['Time']) + current_time_offset
            
            concatenated_data = pd.concat([concatenated_data, sim_data], ignore_index=True)
            
            current_time_offset += pd.Timedelta(1, 'D')
        
        plt.figure(figsize=(14, 6))
        plt.plot(concatenated_data['CGM'], label='CGM', color='orange', linewidth=0.8)
        plt.xlabel('Temps')
        plt.ylabel('Glucosa')
        plt.title(f'Concatenaci√≥ de Simulacions per {adult} - {meal}')
        plt.legend()
        plt.grid(True)
        plt.show()
```
